```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Penguins Data.csv')
df.head()
```

```python
# Crteating a histogram
plt.hist(df['Culmen Length (mm)'], color='red', bins=15)
plt.hist(df['Culmen Depth (mm)'], color='yellow', bins=15)
plt.hist(df['Flipper Length (mm)'], color='green', bins=15)
plt.hist(df['Body Mass (g)'], color='blue', bins=15)
plt.title("Distribution - Penguins Data")
plt.xlabel("Penguins Data")
plt.ylabel("Frequency")
plt.show()
```
