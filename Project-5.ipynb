```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading dataset
df = pd.read_csv('Penguins Data.csv')
df.head()
```

```python
# Creating a scatter plot between Culmen Lentgh and Body Mass
plt.scatter(df['Culmen Length (mm)'], df['Body Mass (g)'], c='pink')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Body Mass (g)')
plt.legend()
plt.show()
```

```python
# Creating a scatter plot between Culmen epth and Body Mass
plt.scatter(df['Culmen Depth (mm)'], df['Body Mass (g)'], c='red')
plt.xlabel('Culmen Depth (mm)')
plt.ylabel('Body Mass (g)')
plt.legend()
plt.show()
```

```python
# Creating a pair plot to check the relationship between all the features of the dataset
pd.plotting.scatter_matrix(df)
plt.show()
```

```python
# Creating an area graph between Culmen Lentgh and Body Mass
plt.stackplot(df['Culmen Length (mm)'], df['Body Mass (g)'], colors=['purple'])
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```
